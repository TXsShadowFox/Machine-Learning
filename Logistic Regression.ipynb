```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "/Users/Downloads/height_weight.csv"
df = pd.read_csv('height_weight.csv')

# Select only Height and Weight columns
df = df[["Height(Inches)", "Weight(Pounds)"]]
df.columns = ["Height", "Weight"]

# Extract values
X = df["Height"].values
y = df["Weight"].values

# Calculate regression coefficients (slope and intercept)
N = len(X)
sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy = np.sum(X * y)
sum_x2 = np.sum(X ** 2)

m = (N * sum_xy - sum_x * sum_y) / (N * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / N

print(f"Linear Regression Equation: y = {m:.2f}x + {c:.2f}")

# Plot the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", alpha=0.6, label="Data Points")
plt.plot(X, m * X + c, color="red", label=f"Regression Line (y = {m:.2f}x + {c:.2f})")
plt.title("Height vs. Weight Linear Regression (Manual Calculation)")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
rmse = 0
for i in range(N):
    y_pred = c + m * X[i]
    rmse += (y[i] - y_pred) ** 2
rmse = np.sqrt(rmse / N)
print(f" (RMSE): {rmse:.2f}")
```

```python
y_pred = c + m * X

# Calculate R-squared for the first 10 rows
ss_res = np.sum((y - y_pred) ** 2)  # Residual sum of squares
ss_tot = np.sum((y - np.mean(y)) ** 2)  # Total sum of squares
r_squared = 1 - (ss_res / ss_tot)  # R-squared

print(f"R-squared (R²) for first 10 rows: {r_squared:.4f}")
```
