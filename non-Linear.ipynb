```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
df = pd.read_csv('employee.csv')


df_clean = df.dropna(subset=["Experience_Years", "Salary"])
X = df_clean["Experience_Years"].values
Y = df_clean["Salary"].values


def exp_model(x, a, b):
    return a * np.exp(b * x)


params, covariance = curve_fit(exp_model, X, Y, p0=(50000, 0.05))
X_sorted = np.linspace(min(X), max(X), 100)
Y_pred = exp_model(X_sorted, *params)


plt.scatter(X, Y, label="Actual Data", color="blue")
plt.plot(X_sorted, Y_pred, label="Non-Linear Fit", color="red")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Non-Linear Regression: Experience vs. Salary")
plt.legend()
plt.show()
print(f"Fitted Parameters: a = {params[0]}, b = {params[1]}")
```
